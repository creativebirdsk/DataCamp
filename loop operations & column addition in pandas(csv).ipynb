```python
mlb= pd.read_csv(r'C:\Users\Dell\Documents\MLB_baseball.csv')
print(mlb)
```

```python
for k,v in mlb.iterrows():     #returns the rows of the databse(csv)
    print(k)
    print(v,'\n')
```

```python
for k,v in mlb.iteritems():   #returns the column items of database(csv)
    print(k)
    print(v)
```

```python
for v in mlb.itertuples():      #returns a single tuple of database(csv)
    #print(k)
    print(v)
```

```python
for k,v in mlb.iterrows():
    print(v['Name']," is a ",v['Position'],' having height = ',v['Height'],' and weight = ',v['Weight'],"\n")
```

```python
for lab,row in mlb.iterrows():     #addition of a row using for loop for calculating name-length using "len()" function
    mlb.loc[lab,'Name_length_FOR']=len(row['Name'])
    print(lab)
    print(row,'\n')
```

```python
for lab,row in mlb.iterrows():     #addition of a row using for loop for calculating capital-name using ".upper()" method
    mlb.loc[lab,'Capital_name_FOR']=row['Name'].upper()
    print(lab)
    print(row,'\n')
```

```python
mlb['Name_length']=mlb['Name'].apply(len)         #works without for loop
for k,v in mlb.iterrows():     #returns the rows of the databse(csv)
    print(k)
    print(v,'\n')
```

```python
mlb['Capital_name']=mlb['Name'].apply(str.upper)      #works without for loop
for k,v in mlb.iterrows():     #returns the rows of the databse(csv)
    print(k)
    print(v,'\n')
```
