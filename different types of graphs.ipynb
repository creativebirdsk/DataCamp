```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year=[1950,1960,1970,1980,1990,2000,2010,2020,2030]
population=[2.5,3.1,4,4.7,5.1,5.9,6.5,7.3,8.1]
np_gdp=np.array([3.5,2,6.4,4.1,9,5.2,4.1,29,74])
np_gdp=np_gdp*6
plt.scatter(year,population,s=np_gdp,c='r')
plt.xlabel('Year')
plt.ylabel('Population [in Billions]')
plt.title('World population graph')
plt.yticks([0,2,4,6,8,10],['0B','2B','4B','6B','8B','10B'])
plt.show()
```
